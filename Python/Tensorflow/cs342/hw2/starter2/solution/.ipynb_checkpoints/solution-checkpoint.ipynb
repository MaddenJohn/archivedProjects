## Part 0: Setup

```python
import tensorflow as tf
import numpy as np
import util

I = tf.placeholder(tf.float32, (None,2), name='input')

# Give the datapoints a label
labels = tf.cast(tf.reduce_sum(I*I, axis=1, keep_dims=True) < 1.0, tf.float32)

RUN_LINEAR = True
RUN_NONLINEAR = True
```

## Part 1: Linear Classifier

```python
if RUN_LINEAR:
    # Step 1: define the compute graph of your linear classifier

    ### Construct your network ###
    weights = tf.get_variable("linear_weights", (2, 1), initializer=tf.random_normal_initializer())
    bias = tf.get_variable("linear_bias", (1, 1), initializer=tf.zeros_initializer())
    logits = tf.add(tf.matmul(I, weights), bias)
    
    linear_output = tf.identity(logits, name='linear_output')

    # Step 2: use a loss function over your classifier's predictions and the ground truth.
    linear_loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(labels=labels, logits=linear_output))

    # Step 3: create an optimizer
    linear_optimizer = tf.train.MomentumOptimizer(0.1, 0.9)

    # Step 4: use that optimizer on your loss function
    linear_mini = linear_optimizer.minimize(linear_loss)
    
    linear_accuracy = tf.reduce_mean(tf.cast(tf.equal(labels, tf.cast(linear_output > 0, tf.float32)), tf.float32))
```

## Part 2: Non-Linear Classifier

```python
if RUN_NONLINEAR:
    # Step 1: define the compute graph of your non-linear classifier
    # Note: use the same input as your linear classifier, python variable 'I' with TF name 'input'.
    #       You will have to branches off the input, your linear and non-linear classifiers

    ### Construct your network ###
    # Hidden non-linear layer
    hidden_weights = tf.get_variable("hidden_weights", (2, 50), initializer=tf.random_normal_initializer())
    hidden_bias = tf.get_variable("hidden_bias", (1, 50), initializer=tf.zeros_initializer())
    hidden_act = tf.nn.relu(tf.add(tf.matmul(I, hidden_weights), hidden_bias))
    
    # Output layer
    out_weights = tf.get_variable("output_weights", (50, 1), initializer=tf.random_normal_initializer())
    out_bias = tf.get_variable("output_bias", (1, 1), initializer=tf.zeros_initializer())
    logit = tf.add(tf.matmul(hidden_act, out_weights), out_bias)
    
    nonlinear_output = tf.identity(logit, name='nonlinear_output')

    # Step 2: use a loss function over your classifier's predictions and the ground truth.
    nonlinear_loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(labels=labels, logits=nonlinear_output))

    # Step 3: create an optimizer
    nonlinear_optimizer = tf.train.MomentumOptimizer(0.1, 0.9)

    # Step 4: use that optimizer on your loss function
    nonlinear_mini = nonlinear_optimizer.minimize(nonlinear_loss)
    
    nonlinear_accuracy = tf.reduce_mean(tf.cast(tf.equal(labels, tf.cast(nonlinear_output > 0, tf.float32)), tf.float32))
```

## Part 3: Training

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

if RUN_LINEAR:
    num_samples = 100
    iterations = 10000
    for i in range(iterations):
        input_data = np.random.rand(num_samples, 2)
        loss_val, _ = sess.run([linear_loss, linear_mini], feed_dict={I: input_data})
    linear_acc_val = sess.run([linear_accuracy], feed_dict={I: input_data})
    print('Final linear model accuracy: ' + str(linear_acc_val))
        
if RUN_NONLINEAR:
    num_samples = 100
    iterations = 10000
    for i in range(iterations):
        input_data = np.random.rand(num_samples, 2)
        loss_val, _ = sess.run([nonlinear_loss, nonlinear_mini], feed_dict={I: input_data})
    nonlinear_acc_val = sess.run([nonlinear_accuracy], feed_dict={I: input_data})
    print('Final non-linear model accuracy: ' + str(nonlinear_acc_val))
```

## Part 4: Visualization

### See your model

```python
# Show the current graph
util.show_graph(tf.get_default_graph().as_graph_def())
```

### See Learned Functions

```python
% matplotlib inline
import matplotlib.pyplot as plt

# Create sample data
num_viz_samples = 10000
viz_data = np.random.rand(num_viz_samples, 2)

labels_outputs = sess.run(labels, feed_dict={I: viz_data})

plt.xlim(0, 1)
plt.ylim(0, 1)

plt.title('Ground Truth')
    
colors = ['black' if l == 0 else 'orange' for l in labels_outputs]
plt.scatter(viz_data[:,0], viz_data[:,1], color=colors)
plt.show()
```

```python
#################################
### Linear Classifier Outputs ###
#################################
if RUN_LINEAR:
    linear_model_predictions = sess.run(linear_output, feed_dict={I: viz_data})
    linear_model_predictions = (linear_model_predictions > 0).astype(int)

    # Plot it
    colors = ['black' if l == 0 else 'orange' for l in linear_model_predictions]
    plt.title('Linear Model')
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.scatter(viz_data[:,0], viz_data[:,1], color=colors)
    plt.show()
```

```python
#####################################
### Non-Linear Classifier Outputs ###
#####################################
if RUN_NONLINEAR:
    # Run it through the classifiers
    nonlinear_model_predictions = sess.run(nonlinear_output, feed_dict={I: viz_data})
    nonlinear_model_predictions = (nonlinear_model_predictions > 0).astype(int)

    # Plot it
    colors = ['black' if l == 0 else 'orange' for l in nonlinear_model_predictions]
    plt.title('Non-linear Model')
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.scatter(viz_data[:,0], viz_data[:,1], color=colors)
    plt.show()
```
