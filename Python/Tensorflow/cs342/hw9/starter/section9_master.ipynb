```python
import tensorflow as tf
from tensorflow.contrib.rnn import RNNCell
import numpy as np
import matplotlib.pyplot as plt
import util
```

```python
% matplotlib inline

# Let's create some sine waves that we can use as our time series data
num_waves = 100
num_samples = 1000
start = 0
end = 50

input_data = np.zeros((num_waves, num_samples))
for w in range(num_waves):
    xs = np.linspace(start, end, num_samples)
    ys = np.sin(xs) * (np.random.rand(1) ** 2)
    input_data[w] = ys

input_data = np.expand_dims(input_data, axis=-1)
print(input_data.shape)
label_data = np.concatenate([input_data[:,1:], np.zeros((num_waves, 1, 1))], axis=1)
print(label_data.shape)
plt.plot(xs, ys)
```

## <a href='https://www.tensorflow.org/api_docs/python/tf/contrib/rnn/RNNCell'>RNNCell</a>

```python
tf.reset_default_graph()

inputs = tf.placeholder(tf.float32, (None, None, 1))
labels = tf.placeholder(tf.float32, (None, None, 1))

cell = tf.contrib.rnn.LSTMCell(100)
```

## <a href='https://www.tensorflow.org/api_docs/python/tf/nn/dynamic_rnn'>dynamic_rnn</a>

```python
lstm_outputs, state = tf.nn.dynamic_rnn(cell, inputs, dtype=tf.float32)

lstm_outputs = tf.reshape(lstm_outputs, [-1, cell.output_size])

outputs = tf.contrib.layers.fully_connected(inputs=lstm_outputs, num_outputs=1, activation_fn=None)
print(outputs)
```

```python
shaped_labels = tf.reshape(labels, [-1, 1])
loss = tf.reduce_mean((outputs - shaped_labels) ** 2)
opt = tf.train.AdamOptimizer().minimize(loss)

sess = tf.Session()

sess.run(tf.global_variables_initializer())
sess.run(tf.local_variables_initializer())
epochs = 10
for e in range(epochs):
    l, _ = sess.run([loss, opt], feed_dict={inputs: input_data, labels: label_data})
    if e % 10 == 0:
        print(l)
```

```python
% matplotlib inline
preds = sess.run(outputs, feed_dict={inputs: input_data})
preds = np.reshape(preds, [num_waves, num_samples, 1])
print(preds.shape)

index = 0
plt.plot(xs, preds[index])
plt.plot(xs, label_data[index])
```

```python
util.show_graph(tf.get_default_graph())
```

## <a href='https://www.tensorflow.org/api_docs/python/tf/contrib/rnn/MultiRNNCell'>MultiRNNCell</a>
Use MultiRNNCell to chain multiple RNNCells together.

```python
tf.reset_default_graph()

inputs = tf.placeholder(tf.float32, (None, None, 1))
labels = tf.placeholder(tf.float32, (None, None, 1))
shaped_labels = tf.reshape(labels, [-1, 1])

cell1 = tf.contrib.rnn.LSTMCell(100)
cell2 = tf.contrib.rnn.GRUCell(50)
multi = tf.contrib.rnn.MultiRNNCell([cell1, cell2])

multi_outputs, multi_state = tf.nn.dynamic_rnn(multi, inputs, dtype=tf.float32)

multi_outputs = tf.reshape(multi_outputs, [-1, multi.output_size])

multi_out = tf.contrib.layers.fully_connected(inputs=multi_outputs, num_outputs=1, activation_fn=None)

multi_loss = tf.reduce_mean((multi_out - shaped_labels) ** 2)
multi_opt = tf.train.AdamOptimizer().minimize(multi_loss)

sess = tf.Session()

sess.run(tf.global_variables_initializer())
sess.run(tf.local_variables_initializer())
epochs = 10
for e in range(epochs):
    l, _ = sess.run([multi_loss, multi_opt], feed_dict={inputs: input_data, labels: label_data})
    if e % 10 == 0:
        print(l)
```

```python
% matplotlib inline
preds = sess.run(multi_out, feed_dict={inputs: input_data})
preds = np.reshape(preds, [num_waves, num_samples, 1])
print(preds.shape)

index = 0
plt.plot(xs, preds[index])
plt.plot(xs, label_data[index])
```

## <a href='https://github.com/tensorflow/tensorflow/blob/r1.3/tensorflow/python/ops/rnn_cell_impl.py'>Custom RNNCells</a>

```python
class FunRNNCell(RNNCell):
    def __init__(self, num_units, activation=None, reuse=None):
        super(FunRNNCell, self).__init__(_reuse=reuse)
        self._num_units = num_units
        self._activation = activation

    @property
    def state_size(self):
        return self._num_units

    @property
    def output_size(self):
        return self._num_units

    def call(self, inputs, state):
        """Most basic RNN: output = new_state = act(W * input + U * state + B)."""
        """FunRNNCell: output = act((W * input) * (U * state))"""
        in_out = tf.contrib.layers.fully_connected(inputs, self._num_units, activation_fn=self._activation)
        state_out = tf.contrib.layers.fully_connected(inputs=inputs, num_outputs=self._num_units, activation_fn=self._activation)
        out = tf.multiply(in_out, state_out)
        return out, out

```

```python
tf.reset_default_graph()

inputs = tf.placeholder(tf.float32, (None, None, 1))
labels = tf.placeholder(tf.float32, (None, None, 1))

cell = FunRNNCell(100)

fun_outputs, state = tf.nn.dynamic_rnn(cell, inputs, dtype=tf.float32)

fun_outputs = tf.reshape(fun_outputs, [-1, cell.output_size])

outputs = tf.contrib.layers.fully_connected(inputs=fun_outputs, num_outputs=1, activation_fn=None)
```

```python
shaped_labels = tf.reshape(labels, [-1, 1])
loss = tf.reduce_mean((outputs - shaped_labels) ** 2)
opt = tf.train.AdamOptimizer().minimize(loss)

sess = tf.Session()

sess.run(tf.global_variables_initializer())
sess.run(tf.local_variables_initializer())
epochs = 100
for e in range(epochs):
    l, _ = sess.run([loss, opt], feed_dict={inputs: input_data, labels: label_data})
    if e % 10 == 0:
        print(l)
```

```python
% matplotlib inline
preds = sess.run(outputs, feed_dict={inputs: input_data})
preds = np.reshape(preds, [num_waves, num_samples, 1])
print(preds.shape)

index = 0
plt.plot(xs, preds[index])
plt.plot(xs, label_data[index])
```
